```python
# Import libraries
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# NLTK VADER for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re
```

```python
news_df = pd.read_csv("tweets/toxic_tweets_1.csv")
news_df.head()
```

```python
news_df2 = pd.read_csv("tweets/toxic_tweets_2.csv")
news_df3 = pd.read_csv("tweets/toxic_tweets_3.csv")
news_df4 = pd.read_csv("tweets/toxic_tweets_4.csv")
news_df5 = pd.read_csv("tweets/toxic_tweets_5.csv")
news_df = pd.concat([news_df,news_df2,news_df3,news_df4,news_df5])
news_df.head()
```

```python
news_df = news_df.sample(frac=1).reset_index(drop=True)
news_df.head(20)
```

```python
news_df.set_index('id', inplace=True)
news_df.to_csv('tweets/toxic_tweets_combined.csv')
```

```python
#Adding some new words to Vader Dictionary to judge stock market news better.

analyser = SentimentIntensityAnalyzer()

scores = news_df['tweet'].apply(analyser.polarity_scores).tolist()
ids = news_df.index.values.tolist()
print(len(scores))
count = 1
for i in scores:
    print("count" + str(count) + str(i))
    count+= 1
scores_df = pd.DataFrame(scores)
scores_df['id'] = ids
scores_df.set_index('id', inplace=True)
scores_df.head()
```

```python
df = pd.concat([news_df,scores_df],axis=1)
df.head(30)
```

```python
df.to_csv('tweets/tweets_with_scores.csv')
```

```python
news_df = pd.read_csv("tweets/tweets_with_scores.csv")
news_df = news_df.iloc[:3000]
news_df.head()
```

```python
score = news_df['compound'].values
toxic = []
severe_toxic = []

for i in score:
    if i < -0.5:
        toxic.append(1)
        severe_toxic.append(1)
    elif i>=-0.5 and i <0:
        toxic.append(1)
        severe_toxic.append(0)
    else:
        toxic.append(0)
        severe_toxic.append(0)
```

```python
news_df['toxic'] = toxic
news_df['severe_toxic'] = severe_toxic
news_df.head()
```

```python
news_df.set_index('id', inplace=True)
news_df.to_csv('tweets/tweets_with_labels.csv')
```
