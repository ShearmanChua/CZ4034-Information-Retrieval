```python
import os
import tweepy as tw
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv("negative_tweets.csv")
df.head()
```

```python
geolocator = Nominatim(user_agent="toxic_tweet_analyser")

from geopy.extra.rate_limiter import RateLimiter

geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df['location'] = df['user_location'].apply(geocode)

df['user_location'] = df['location'].apply(lambda loc: str(loc.address) if loc else None)
df['user_geo'] = df['location'].apply(lambda loc: str(loc.point) if loc else None)
```

```python
df['user_geo'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
df.head(20)
```

```python
df = df[df.user_location.notnull()]
df.info()
```

```python
df.to_csv('negative_tweets_loc.csv',index = False,encoding='utf-8')
```
